```python
from ultralytics import YOLO
from ultralytics import SAM

# Load YOLO model
model = YOLO("runs/detect/train/weights/best.pt")  # Pretrained YOLO model

# Run batched inference on a list of images
results = model("test_images/meningioma_3.jpg", save=True)  # returns a list of Results objects

# Load the SAM model
sam_model = SAM("sam2_b.pt")  # Use variant sam2_b: SAM base model

for result in results:
    class_ids = result.boxes.cls.int().tolist()
    if len(class_ids):
        boxes = result.boxes.xyxy  # Boxes object for bounding box outputs
        sam_results = sam_model(result.orig_img, bboxes=boxes, verbose=False, save=True, device="cpu")
```
